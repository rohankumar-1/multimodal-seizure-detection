```python
import numpy as np
import pandas as pd
```

```python
train_8Hz = np.load("../data/processed_2s_8Hz/train.npz", allow_pickle=True)
val_8Hz = np.load("../data/processed_2s_8Hz/val.npz", allow_pickle=True)
test_8Hz = np.load("../data/processed_2s_8Hz/test.npz", allow_pickle=True)

train_40Hz = np.load("../data/processed_2s_40Hz/train.npz", allow_pickle=True)
val_40Hz = np.load("../data/processed_2s_40Hz/val.npz", allow_pickle=True)
test_40Hz = np.load("../data/processed_2s_40Hz/test.npz", allow_pickle=True)

train_128Hz = np.load("../data/processed_2s_128Hz/train.npz", allow_pickle=True)
val_128Hz = np.load("../data/processed_2s_128Hz/val.npz", allow_pickle=True)
test_128Hz = np.load("../data/processed_2s_128Hz/test.npz", allow_pickle=True)
```

```python
array_dict = {
    "8Hz": {
        "train": train_8Hz,
        "val": val_8Hz,
        "test": test_8Hz,
    },
    "40Hz": { 
        "train": train_40Hz,
        "val": val_40Hz,
        "test": test_40Hz,
    },
    "128Hz": {
        "train": train_128Hz,
        "val": val_128Hz,
        "test": test_128Hz,
    },
}
```

```python
label_value_counts_dict = {}

for fs in array_dict.keys():
    label_value_counts_dict[fs] = {}
    for split in array_dict[fs].keys():
        label_value_counts_dict[fs][split] = {}
        label_value_counts_dict[fs][split]["binary_label"] = pd.Series(array_dict[fs][split]["binary_label"]).value_counts()
        label_value_counts_dict[fs][split]["lateralization"] = pd.Series(array_dict[fs][split]["lateralization"]).value_counts()
        label_value_counts_dict[fs][split]["label"] = pd.Series(array_dict[fs][split]["label"]).value_counts()
        label_value_counts_dict[fs][split]["localization"] = pd.Series(array_dict[fs][split]["localization"]).value_counts()
```
